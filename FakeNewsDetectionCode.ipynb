```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
true_df =  pd.read_csv("True.csv")
true_df
```

```python
fake_df =  pd.read_csv("Fake.csv")
fake_df
```

```python
true_df["class"] = 1
fake_df["class"] = 0
```

```python
df = pd.concat([true_df,fake_df],axis =0)
```

```python
df
```

```python
df.drop(["title","subject","date"],axis = 1,inplace = True)
```

```python
df.isnull().sum()
```

## Do random shuffling of data

```python
df = df.sample(frac= 1)
```

```python
df
```

```python
df.reset_index(inplace = True)
```

```python
df.drop(["index"],axis = 1,inplace = True)
df
```

```python
df.columns
```

```python
nltk.download('stopwords')
```

```python
# printing the stopwords in English
print(stopwords.words('english'))
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
df["text"] = df["text"].apply(stemming)
df.head()
```

```python
df.head()
```

```python
X = df["text"]
Y = df["class"]
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
Vectorizer = TfidfVectorizer()
# fit_transform method for training data to both learn the vocabulary and transform the data
XV_train = Vectorizer.fit_transform(X_train)
# transform method is being used with test data to apply the same mapping without altering the learned vocabulary and IDF values.
XV_test = Vectorizer.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(XV_train,Y_train)
```

```python
predict_lr = lr.predict(XV_test)
predict_lr
```

```python
lr.score(XV_test,Y_test)
```

```python
dt = DecisionTreeClassifier()
dt.fit(XV_train,Y_train)
```

```python
predict_dt = dt.predict(XV_test)
```

```python
dt.score(XV_test,Y_test)
```

```python
rec = RandomForestClassifier(random_state = 0)
rec.fit(XV_train,Y_train)
```

```python
predict_rec = rec.predict(XV_test)
```

```python
rec.score(XV_test,Y_test)
```

```python
gb =  GradientBoostingClassifier(random_state =0)
gb.fit(XV_train,Y_train)
```

```python
predict_gb = gb.predict(XV_test)
```

```python
gb.score(XV_test,Y_test)
```
